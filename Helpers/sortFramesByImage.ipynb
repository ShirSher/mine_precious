```python
def contains_square(image_path, min_area=100, epsilon_factor=0.03):

    image = cv2.imread(image_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    edged = cv2.Canny(blurred, 30, 150)
    
    contours, _ = cv2.findContours(edged.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    if(DEBUG_MODE):
        # Draw contours on the image
        cv2.drawContours(image, contours, -1, (0, 255, 0), 2)

        # Display the image using matplotlib
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        plt.axis('off')  # To hide axis labels
        plt.show()

    for cnt in contours:
        # Calculate area and remove small elements
        area = cv2.contourArea(cnt)
        if area < min_area:
            continue
        if(DEBUG_MODE):
            print("area ", area)
        
        approx = cv2.approxPolyDP(cnt, epsilon_factor * cv2.arcLength(cnt, True), True)
        
        if(DEBUG_MODE):
            print(approx)
        
        # If the contour has four vertices, we assume it's a square
        if len(approx) >= 4 and len(approx) <= 10:
            return True

    return False
    
if(DEBUG_MODE):
    print(contains_square(DEBUG_IMAGE))
```

```python
def find_squares(folder_path, starting_idx = 0, threshold=500000):

    image_files = sorted([f for f in os.listdir(folder_path) if f.endswith('.png')])

    strats_ends = [get_num_from_path(image_files[starting_idx])]
    frame_idx = starting_idx
    
    while (frame_idx+10) < len(image_files):

        last_frame = frame_idx
        for comp_idx in range(10):
            idx = comp_idx+frame_idx
            image_file = image_files[idx]
            image_path = os.path.join(folder_path, image_file)
            curr_w_square = contains_square(image_path)
            if (curr_w_square):
                last_frame = idx
        strats_ends.append(get_num_from_path(image_files[last_frame]))

        frame_idx = last_frame + 28
        if ((frame_idx+10) > len(image_files)):
            break

        first_frame = frame_idx
        for comp_idx in range(10):
            idx = comp_idx+frame_idx
            image_file = image_files[idx]
            image_path = os.path.join(folder_path, image_file)
            curr_w_square = contains_square(image_path)
            if (curr_w_square):
                first_frame = idx
                break
        strats_ends.append(get_num_from_path(image_files[first_frame]))
        frame_idx = first_frame + 58

    print(strats_ends)
```
